```python
def arithmetic_arranger(angka):
    if type(angka) == list:
        result = ""
        for i in angka:
            a = i.find("+")
            if a == -1:
                a = i.find("-")
            c = a - 1
            b = a + 2
            num1 = i[:c].strip()
            num2 = i[b:].strip()

            # Menentukan panjang maksimum antara num1 dan num2
            max_length = max(len(num1), len(num2))

            num1 = num1.rjust(max_length)  # Menambahkan spasi ke kiri
            num2 = num2.rjust(max_length)  # Menambahkan spasi ke kiri

            op = i[a]

            s = 2 + max_length
            s1 = s - len(num1)
            s2 = s - len(num2) - 1

            result += num1 + "    "
            result += op + " " * (s2 - 4) + num2 + "    "  # Menambahkan spasi di antara hasil
        return result  # Mengembalikan hasil sebagai string
    else:
        return "Input harus berupa list masalah penjumlahan."

# List masalah yang ingin diolah
mylist = ["32 + 698", "3801 - 2", "45 + 43", "123 + 49"]

# Memanggil fungsi dan mencetak hasil
result = arithmetic_arranger(mylist)
print(result)
```
