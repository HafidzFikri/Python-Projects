```python
Day = ["monday", "tuesday","wednesday","thursday", "friday", "saturday", "sunday"]
def time_calculator(time,plustime,day=""):
  T = time.find(":")
  thour = time[:T].strip()
  thour = int(thour)
  tminute = time[T+1:T+3].strip()
  tminute = int(tminute)
#  thour <= 24 and thour >= 0
#  tminute <= 61 and tminute >= 0
  A = time.find(" ")
  AMPM = time[A+1:].strip()
  if AMPM == "PM":
    hour = thour + 12
  else:
    hour = thour

  PT = plustime.find(":")
  pthour = plustime[:PT].strip()
  pthour = int(pthour)
  ptminute = plustime[PT+1:].strip()
  ptminute = int(ptminute)

  totalminute = tminute + ptminute
  if totalminute >= 60:
    plushour = totalminute // 60
    realminute = totalminute % 60
  else:
    plushour = 0
    realminute = totalminute

  totalhour = hour + pthour + plushour
  if totalhour >= 24:
    realhour = totalhour % 24
    plusday = totalhour // 24
  else:
    realhour = totalhour
    plusday = 0
  realhour = int(realhour)

  if realhour < 12:
    ampm = "AM"
  elif realhour > 11:
    realhour = int(realhour) - 12
    ampm = "PM"

  if day == "":
    D = ""
    DAYY = ""
  else:
    D = Day.index(day.lower())
    dayy = D + plusday
    dayy = int(dayy)
    while dayy > 6:
      dayy = dayy - 6

    DAY = Day[dayy]
    DAYY = ", " + DAY
  if plusday == 0:
    APM = ""
    day = day
  elif plusday == 1:
    c = D + 1
 #   day = Day[c]
    APM = "(Next Day)"
  else:
    APM = f"({plusday} Days Later)"
    #b = D + plusday
#    day = Day[b]


  realhour = str(realhour)
  realminute = str(realminute)
  TIME = f"{realhour}:{realminute} {ampm}{DAYY} {APM}"
  print(TIME)

time_calculator("10:10 AM","90:70")
time_calculator("11:20 AM","90:70","monday")
time_calculator("9:15 AM","4:50", "Saturday")
time_calculator("10:10 AM","2000:20", "TuEsDaY")
```
